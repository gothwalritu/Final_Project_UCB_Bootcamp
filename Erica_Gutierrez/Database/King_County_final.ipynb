```python
# Make a connection to the SQL database
conn = sql.connect('King_County_WQI.db')
```

```python
# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{'King_County_WQI.db'}")
```

```python
# Get the name of the table. 
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Using the inspector to print the column names within the 'King_County_WQI ' table and its types
columns = inspector.get_columns('King_County_WQI ')
for column in columns:
    print(column)
```

```python
# Get the column names for the table
column_names = inspector.get_columns('King_County_WQI')
for column_name in column_names:
    print(column_name["name"])
```

```python
# Commit to flush change to disk, then close connection 
conn.commit()
conn.close()
```

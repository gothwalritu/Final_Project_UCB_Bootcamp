```python
# SQLAlchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
from datetime import datetime 

import pandas as pd 
import sqlite3 as sql
```

```python
file_path="Resources/Big_Lake_Sammamish.csv"
WQ_Big_Lake_df=pd.read_csv(file_path)
WQ_Big_Lake_df.head(10)
```

```python
list(WQ_Big_Lake_df.columns)
```

```python
WQ_Big_Lake_df.drop('Unnamed: 0',axis=1, inplace=True)
WQ_Big_Lake_df
```

```python
#Rename Ammmonia_Nitrogen to Ammonia_Nitrogen
WQ_Big_Lake_df.rename({'Ammmonia Nitrogen (mg/L)':'Ammonia Nitrogen (mg/L)'},axis='columns',inplace=True)
WQ_Big_Lake_df.head()
```

```python
#Create the database
conn = sql.connect('WQI_King_County.db')
c = conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS Big_Lake_Sammamish (collectdate , Ammonia_Nitrogen , Conductivity , Dissolved_Oxygen , Fecal_Coliform INTEGER, Nitrate_Nitrite , Orthophosphate_Phosphate , pH , Temperature , Total_Nitrogen INTEGER, Total_Phosphorous , WQI)')
conn.commit()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS CollectDate (Ammonia_Nitrogen,Conductivity,Dissolved_Oxygen,Fecal_Coliform, Nitrate_Nitrite ,Orthophosphate_Phosphate,pH,Temperature, Total_Nitrogen,Total_Phosphorous,WQI)')
conn.commit()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS Prediction (Prediction, Actual)')
conn.commit()
```

```python
# Get from Pandas DataFrame to SQL 
WQ_Big_Lake_df.to_sql('Big_Lake_Sammamish', conn, if_exists='replace', index = False)
```

```python
# Get from Pandas DataFrame to SQL 
WQ_Big_Lake_df.to_sql('CollectDate', conn, if_exists='replace', index = False)
```

```python
# Get from Pandas DataFrame to SQL 
WQ_Big_Lake_df.to_sql('Prediction', conn, if_exists='replace', index = False)
```

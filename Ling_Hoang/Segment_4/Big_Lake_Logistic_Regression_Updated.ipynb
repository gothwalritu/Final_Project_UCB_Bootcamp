## Set Up Logistic Regression Model

```python
# Import dependencies for ML
import pandas as pd
from sklearn import utils
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Seperate the Features (X) from the Target (y)
y = Big_Lake_df['WQ']

# Drop 'WQI'/'WQ' because those are outputs
X = Big_Lake_df.drop(['WQI', 'WQ', 'CollectDate'], axis='columns')
```

```python
# Split our data into training and testing 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=78)
                                                    
# Print X_test_shape                                                    
X_test.shape
```

```python
# Print y_train
y_train
```

```python
# Check for y_train value counts in each categorical
y_train.value_counts()
```

```python
# Print y_tes
y_test
```

```python
# Create a Logistic Regression Model 
# For multiclass problems, only ‘newton-cg’, ‘sag’, ‘saga’ and ‘lbfgs’ handle multinomial loss
# 'lbfgs' has error so switching to 'newton-cg'
classifier = LogisticRegression(solver='newton-cg', random_state=0)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
# Convert y values to categorical values
lab = preprocessing.LabelEncoder()
y_transformed = lab.fit_transform(y_train)
```

```python
# View transformed values
print(y_transformed)
```

```python
# Fit (train) or model using the training data including y_trainsformed
classifier.fit(X_train, y_transformed)
```

```python
# Prediction outcomes for test data set 
predictions = classifier.predict(X_test)

results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Print accuracy score 
print(accuracy_score(lab.fit_transform(y_test), predictions))
```

```python
# Print total mean value of coefficient
np.mean(classifier.coef_)
```

```python
# Print mean of absolute value for each parameters
# See the coefficient of each parameter with WQI
np.mean(np.abs(classifier.coef_), 0)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(lab.fit_transform(y_test), predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3", "Actual 4", "Actual 5"],
    columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3", "Predicted 4", "Predicted 5"])

cm_df
```

```python
# Print confusion matrix table 
confusion_matrix(lab.fit_transform(y_test), predictions)
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(lab.fit_transform(y_test), predictions)# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(lab.fit_transform(y_test), predictions))
```
